```python
import pptx
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.enum.text import PP_ALIGN
from pptx.dml.color import RGBColor

def create_presentation():
    # Create presentation object
    presentation = Presentation()

    # Slide 1 - Title Slide
    slide_layout = presentation.slide_layouts[0]
    slide = presentation.slides.add_slide(slide_layout)

    title = slide.shapes.title
    subtitle = slide.placeholders[1]

    title.text = "Sample PowerPoint Presentation"
    subtitle.text = "Created using python-pptx"

    # Slide 2 - Content Slide
    slide_layout = presentation.slide_layouts[1]
    slide = presentation.slides.add_slide(slide_layout)

    title = slide.shapes.title
    content = slide.placeholders[1]

    title.text = "Agenda"

    tf = content.text_frame
    tf.text = "Topics Covered:"

    points = [
        "Introduction",
        "Main Content",
        "Conclusion"
    ]

    for point in points:
        p = tf.add_paragraph()
        p.text = point
        p.level = 1

    # Slide 3 - Custom Formatting
    slide_layout = presentation.slide_layouts[5]
    slide = presentation.slides.add_slide(slide_layout)

    left = Inches(1)
    top = Inches(2)
    width = Inches(8)
    height = Inches(1.5)

    textbox = slide.shapes.add_textbox(left, top, width, height)
    tf = textbox.text_frame

    p = tf.paragraphs[0]
    run = p.add_run()
    run.text = "Thank You!"
    run.font.size = Pt(40)
    run.font.bold = True
    run.font.italic = True
    run.font.color.rgb = RGBColor(0, 0, 255)
    p.alignment = PP_ALIGN.CENTER

    # Save presentation
    presentation.save("sample_presentation.pptx")
    print("PowerPoint file created successfully!")

if __name__ == "__main__":
    create_presentation()
```
